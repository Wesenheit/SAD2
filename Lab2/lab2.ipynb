```python
import numpy as np
import scipy.stats as s
import matplotlib.pyplot as plt
```

 # Berkson's paradox

## Draw 1000 samples of $x$ and $y$ from uniform distribution and a) calculate Pearson correlation, b) Plot scatterplot

```python
n=1000
x=np.random.rand(n)
y=np.random.rand(n)
plt.scatter(x,y)
t=s.pearsonr(x,y)
print(t)
```

## Condition dataset on the collider: $x$ + $y$ > 0.2  and calculate Pearson correlation, and a) calculate Pearson correlation, b) Plot scatterplot


```python
var=(x+y>0.2)
x_p=x[var]
y_p=y[var]
plt.scatter(x_p,y_p)
t=s.pearsonr(x_p,y_p)
print(t)
t.pvalue
```

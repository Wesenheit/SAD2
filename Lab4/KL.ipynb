```python
import torch
from torch import distributions as tcd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex']=True
sns.set_theme()
```

```python
mean2=0
sigma2=1
mean1_=np.linspace(-5,5,100)
sigma1_=np.linspace(0.001,10,100)
mean1,sigma1=np.meshgrid(mean1_,sigma1_)
div=np.log(sigma1/sigma2)+(sigma1**2+(mean1-mean2)**2)/(2*sigma2**2)-1/2
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(mean1,sigma1,div,color="red")
ax.set_xlabel(r"$\mu$")
ax.set_ylabel(r"$\sigma$")
ax.view_init(30,-70)
```
